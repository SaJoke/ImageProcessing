# Setup parameters

```python
img1 = circle_small
img2 = circle_big
img3 = ellipse
img4 = Z_shape
img5 = rectangle
img6 = hand_small
img7 = hand_big

ref_img = np.float32(img6)/255
temp_img = np.float32(img7)/255
m,n = ref_img.shape

#common parameters
Imax = 1000
alpha = 1e-1
ssd_eps = 5e-2
res_eps = 5e-2
diff_eps = 1e-4
print_every = 50

#Time Marching paremeter
tau = 5e-1

#Fixed Point
SORmax = 4
```

# Setup tools

```python
#used in 3d plot
X = np.arange(0,m,1)
Y = np.arange(0,n,1)
X, Y = np.meshgrid(X, Y)

def ex_neuBCs(u): #extended u by neuman BCs
    ex_u = np.zeros((2,m+2,n+2))
    ex_u[:,1:-1,1:-1] = u
    ex_u[:,0,1:-1] = u[:,0,:]
    ex_u[:,1:-1,0] = u[:,:,0]
    ex_u[:,-1,1:-1] = u[:,-1,:]
    ex_u[:,1:-1,-1] = u[:,:,-1]
    return ex_u

def lap(ex_u):
    #laplacian (4 bounded sides are deleted)
    return ex_u[:,:-2,1:-1]+ex_u[:,1:-1,:-2]+ex_u[:,2:,1:-1]+ex_u[:,1:-1,2:]-4*ex_u[:,1:-1,1:-1]

def cdiff(ex_Tu):
    Tu_dx = (ex_Tu[2:,1:-1]-ex_Tu[:-2,1:-1])/2 #central
    Tu_dy = (ex_Tu[1:-1,2:]-ex_Tu[1:-1,:-2])/2 #central
    Tu_du = np.zeros((2,m,n))
    Tu_du[0,:,:] = Tu_dx
    Tu_du[1,:,:] = Tu_dy
    return Tu_du

def show_result(ref_img,regist_Diff_ETM_img,temp_img,grid_warp,vec_field,X,Y,u):
    fig = plt.figure(figsize=(15,15))
    fig.add_subplot(3,3,1)
    plt.title('Reference Image')
    plt.imshow(ref_img,cmap='gray')
    fig.add_subplot(3,3,2)
    plt.title('Registed Image')
    plt.imshow(regist_Diff_ETM_img,cmap='gray')
    fig.add_subplot(3,3,3)
    plt.title('Template Image')
    plt.imshow(temp_img,cmap='gray')
    fig.add_subplot(3,3,4)
    plt.title('Grid warp')
    plt.imshow(grid_warp,cmap='gray')
    fig.add_subplot(3,3,5)
    plt.title('Vector Field')
    plt.imshow(vec_field,cmap='gray')
    fig.add_subplot(3,3,6)
    plt.title('relative SSD - iteration')
    plt.scatter(np.arange(k+1),ssd_list,color="green",marker="*",s=10)
    u_1 = fig.add_subplot(337,projection='3d')
    plt.title('u_1')
    u_1.plot_surface(X, Y, u[0,:,:], cmap=cm.coolwarm,linewidth=0, antialiased=False)
    u_2 = fig.add_subplot(338,projection='3d')
    plt.title('u_2')
    u_2.plot_surface(X, Y, u[1,:,:], cmap=cm.coolwarm,linewidth=0, antialiased=False)
    plt.show()
    return None
    
```

# Explicit Time Marching

```python
t_start = time.time()
u = np.zeros((2,m,n)) #initial solution
ssd_list = [] #used in relative ssd - iteration plot
res0_exist = False #used to find relative ssd
for k in range(Imax+1):
    u_old = np.copy(u)
    ex_u = ex_neuBCs(u)
    
    Lp_u = lap(ex_u) #laplacian of u
    
    ex_Tu = GrayImg.warp_vec(temp_img,ex_u,up=1,left=1,right=1,down=1)
    Tu = ex_Tu[1:-1,1:-1]
    Tu_du = cdiff(ex_Tu) #central approximated
    fu = (Tu-ref_img)*Tu_du
    
    Res = alpha*Lp_u-fu #residue array
    
    norm_Res = np.linalg.norm(Res)
    ssd = (np.sum((Tu-ref_img)**2))/2
    if not res0_exist:
        res0 = norm_Res
        ssd0 = ssd
        res0_exist = True
    rel_res = norm_Res/res0
    rel_ssd = ssd/ssd0
    ssd_list.append(rel_ssd)
    if k%print_every==0:
        t_loop = time.time()
        print('\trel_res =',rel_res,'\n\trel_ssd =',rel_ssd)
    if rel_res<res_eps or rel_ssd<ssd_eps:
        print('solution converges')
        break
        
    if k<Imax:
        u = u+tau*Res #update u
        
        rel_diff = np.linalg.norm(u-u_old)/np.linalg.norm(u)
        if (k+1)%print_every==0:
            print(k+1)
            print('\t',print_every,'rounds take time =',time.time()-t_loop)
            print('\trel_diff = ',rel_diff)
        if rel_diff<diff_eps:
            print('solution dose not change')
            break        
if k==Imax:
    print('Exceed')
regist_Diff_ETM_img = np.clip(Tu,0,1)
grid_array = 1-GrayImg.grid_array((m,n),30,30)
grid_warp = GrayImg.warp_vec(grid_array,u) #bilinear interpolation
vec_field = GrayImg.vec_field_array((m,n),u,30,30)

#show results
print('\n\niteration =',k,'\nrel_diff = ',rel_diff,'\nrel_res =',
      rel_res,'\nrel_ssd =',rel_ssd,'\ntotal_time =',time.time()-t_start)
show_result(ref_img,regist_Diff_ETM_img,temp_img,grid_warp,vec_field,X,Y,u)
```

# Semi-Explicit Time Marching

```python
from scipy import sparse
from scipy.sparse.linalg import inv

#find inv(I+alpha*tau*A)
t_inv = time.time()
K = GrayImg.diffusion_array((m,n),alpha*tau,tosparse=True)
invK = inv(sparse.csc_matrix(K))
print('inv time =',time.time()-t_inv)
```

```python
t_start = time.time()
u = np.zeros((2,m,n)) #initial solution
ssd_list = [] #used in relative ssd - iteration plot
res0_exist = False #used to find relative ssd

for k in range(Imax+1):
    u_old = np.copy(u)
    ex_u = ex_neuBCs(u)
    
    Lp_u = lap(ex_u) #laplacian of u
    
    ex_Tu = GrayImg.warp_vec(temp_img,ex_u,up=1,left=1,right=1,down=1)
    Tu = ex_Tu[1:-1,1:-1]
    Tu_du = cdiff(ex_Tu) #central approximated
    fu = (Tu-ref_img)*Tu_du
    
    Res = alpha*Lp_u-fu #residue array
    
    norm_Res = np.linalg.norm(Res)
    ssd = (np.sum((Tu-ref_img)**2))/2
    if not res0_exist:
        res0 = norm_Res
        ssd0 = ssd
        res0_exist = True
    rel_res = norm_Res/res0
    rel_ssd = ssd/ssd0
    ssd_list.append(rel_ssd)
    if k%print_every==0:
        t_loop = time.time()
        print('\trel_res =',rel_res,'\n\trel_ssd =',rel_ssd)
    if rel_res<res_eps or rel_ssd<ssd_eps:
        print('solution converges')
        break
        
    if k<Imax:
        #update u
        F = (u-tau*fu).reshape(2,m*n)
        Fx = invK.dot(F[0])
        Fy = invK.dot(F[1])
        u[0] = Fx.reshape(m,n)
        u[1] = Fy.reshape(m,n)
        
        rel_diff = np.linalg.norm(u-u_old)/np.linalg.norm(u)
        if (k+1)%print_every==0:
            print(k+1)
            print('\t',print_every,'rounds take time =',time.time()-t_loop)
            print('\trel_diff = ',rel_diff)
        if rel_diff<diff_eps:
            print('solution dose not change')
            break        
if k==Imax:
    print('Exceed')
regist_Diff_ETM_img = np.clip(Tu,0,1)
grid_array = 1-GrayImg.grid_array((m,n),30,30)
grid_warp = GrayImg.warp_vec(grid_array,u) #bilinear interpolation
vec_field = GrayImg.vec_field_array((m,n),u,30,30)

#show results
print('\n\niteration =',k,'\nrel_diff = ',rel_diff,'\nrel_res =',
      rel_res,'\nrel_ssd =',rel_ssd,'\ntotal_time =',time.time()-t_start)
show_result(ref_img,regist_Diff_ETM_img,temp_img,grid_warp,vec_field,X,Y,u)
```

# AOS

```python
from scipy import sparse
from scipy.sparse.linalg import inv

#find inv(I+alpha*tau*A) as invK
t_inv = time.time()
K2 = GrayImg.diffusion_array((m,n),2*alpha*tau,tosparse=True)
dKx = K2.diagonal(0).copy()
dKx[1:] += K2.diagonal(1)
dKx[:-1] += K2.diagonal(-1)
dKy = K2.diagonal(0)-dKx+1
datax = [K2.diagonal(-n),dKx,K2.diagonal(n)]
datay = [K2.diagonal(-1),dKy,K2.diagonal(1)]
Kx = sparse.diags(datax,[-n,0,n])
Ky = sparse.diags(datay,[-1,0,1])
invKx = inv(sparse.csc_matrix(Kx))
invKy = inv(sparse.csc_matrix(Ky))
approx_invK = (invKx+invKy)/2
print('inv time =',time.time()-t_inv)
```

```python
t_start = time.time()
u = np.zeros((2,m,n)) #initial solution
ssd_list = [] #used in relative ssd - iteration plot
res0_exist = False #used to find relative ssd

for k in range(Imax+1):
    u_old = np.copy(u)
    ex_u = ex_neuBCs(u)
    
    Lp_u = lap(ex_u) #laplacian of u
    
    ex_Tu = GrayImg.warp_vec(temp_img,ex_u,up=1,left=1,right=1,down=1)
    Tu = ex_Tu[1:-1,1:-1]
    Tu_du = cdiff(ex_Tu) #central approximated
    fu = (Tu-ref_img)*Tu_du
    
    Res = alpha*Lp_u-fu #residue array
    
    norm_Res = np.linalg.norm(Res)
    ssd = (np.sum((Tu-ref_img)**2))/2
    if not res0_exist:
        res0 = norm_Res
        ssd0 = ssd
        res0_exist = True
    rel_res = norm_Res/res0
    rel_ssd = ssd/ssd0
    ssd_list.append(rel_ssd)
    if k%print_every==0:
        t_loop = time.time()
        print('\trel_res =',rel_res,'\n\trel_ssd =',rel_ssd)
    if rel_res<res_eps or rel_ssd<ssd_eps:
        print('solution converges')
        break
        
    if k<Imax:
        #update u
        F = (u-tau*fu).reshape(2,m*n)
        Fx = approx_invK.dot(F[0])
        Fy = approx_invK.dot(F[1])
        u[0] = Fx.reshape(m,n)
        u[1] = Fy.reshape(m,n)
        
        rel_diff = np.linalg.norm(u-u_old)/np.linalg.norm(u)
        if (k+1)%print_every==0:
            print(k+1)
            print('\t',print_every,'rounds take time =',time.time()-t_loop)
            print('\trel_diff = ',rel_diff)
        if rel_diff<diff_eps:
            print('solution dose not change')
            break        
if k==Imax:
    print('Exceed')
regist_Diff_ETM_img = np.clip(Tu,0,1)
grid_array = 1-GrayImg.grid_array((m,n),30,30)
grid_warp = GrayImg.warp_vec(grid_array,u) #bilinear interpolation
vec_field = GrayImg.vec_field_array((m,n),u,30,30)

#show results
print('\n\niteration =',k,'\nrel_diff = ',rel_diff,'\nrel_res =',
      rel_res,'\nrel_ssd =',rel_ssd,'\ntotal_time =',time.time()-t_start)
show_result(ref_img,regist_Diff_ETM_img,temp_img,grid_warp,vec_field,X,Y,u)
```

# Fixed point

```python
from scipy import sparse

fp_Imax = 20
fp_print_every = 1
al_lp = -GrayImg.diffusion_array((m,n),alpha,tosparse=True)
al_lp.setdiag(al_lp.diagonal(0)+1, k=0)
```

```python
t_start = time.time()
u = np.zeros((2,m,n)) #initial solution
ssd_list = [] #used in relative ssd - iteration plot
res0_exist = False #used to find relative ssd
for k in range(fp_Imax+1):
    u_old = np.copy(u)
    ex_u = ex_neuBCs(u)
    
    Lp_u = lap(ex_u) #laplacian of u
    
    ex_Tu = GrayImg.warp_vec(temp_img,ex_u,up=1,left=1,right=1,down=1)
    Tu = ex_Tu[1:-1,1:-1]
    Tu_du = cdiff(ex_Tu) #central approximated
    fu = (Tu-ref_img)*Tu_du
    
    Res = alpha*Lp_u-fu #residue array
    
    norm_Res = np.linalg.norm(Res)
    ssd = (np.sum((Tu-ref_img)**2))/2
    if not res0_exist:
        res0 = norm_Res
        ssd0 = ssd
        res0_exist = True
    rel_res = norm_Res/res0
    rel_ssd = ssd/ssd0
    ssd_list.append(rel_ssd)
    if k%fp_print_every==0:
        t_loop = time.time()
        print('\trel_res =',rel_res,'\n\trel_ssd =',rel_ssd)
    if rel_res<res_eps or rel_ssd<ssd_eps:
        print('solution converges')
        break
        
    if k<fp_Imax:
        #update u
        sigma = Tu_du**2
        G = -fu+sigma*u
        A1 = sparse.diags(sigma[0].reshape(m*n))-al_lp
        A2 = sparse.diags(sigma[0].reshape(m*n))-al_lp
        su1 = LinSys.SOR(A1,G[0].reshape(m*n),x0=u[0].reshape(m*n),Imax=SORmax,eps=1e-4,weight=1,progress=False)
        su2 = LinSys.SOR(A2,G[1].reshape(m*n),x0=u[1].reshape(m*n),Imax=SORmax,eps=1e-4,weight=1,progress=False)
        u[0] = su1.reshape(m,n)
        u[1] = su2.reshape(m,n)
        
        rel_diff = np.linalg.norm(u-u_old)/np.linalg.norm(u)
        if (k+1)%fp_print_every==0:
            print(k+1)
            print('\t',fp_print_every,'rounds take time =',time.time()-t_loop)
            print('\trel_diff = ',rel_diff)
        if rel_diff<diff_eps:
            print('solution dose not change')
            break        
if k==fp_Imax:
    print('Exceed')
regist_Diff_ETM_img = np.clip(Tu,0,1)
grid_array = 1-GrayImg.grid_array((m,n),30,30)
grid_warp = GrayImg.warp_vec(grid_array,u) #bilinear interpolation
vec_field = GrayImg.vec_field_array((m,n),u,30,30)

#show results
print('\n\niteration =',k,'\nrel_diff = ',rel_diff,'\nrel_res =',
      rel_res,'\nrel_ssd =',rel_ssd,'\ntotal_time =',time.time()-t_start)
show_result(ref_img,regist_Diff_ETM_img,temp_img,grid_warp,vec_field,X,Y,u)
```
