```python
import cv2,numpy
from matplotlib import pyplot as plt
from JokeFunc import GrayImg
from importlib import reload
GrayImg = reload(GrayImg)
```

```python
#create a rectangle
height,width = 480,720
image = 100*numpy.ones((height,width), numpy.uint8)
image = cv2.rectangle(image, (width//4,height//4), (3*width//4,3*height//4), color=(255,255,255), thickness=-1)
   
# Displaying the image
plt.figure(figsize=(10,5))
plt.imshow(image , cmap='gray', vmin=0, vmax=255)
cv2.imwrite('Image\TemplateImg.png',image)
```

# Warp Affine with bilinear interpolation

```python
from math import floor,ceil,sqrt,cos,sin,pi

#Rotation
theta = 0
Ar = numpy.array([[cos(theta),-sin(theta)],[sin(theta),cos(theta)]]) 

#Scaling
s1 = 2
s2 = 2
Asc = numpy.array([[s1,0],[0,s2]])

#Shearing
S = 1
Ash = numpy.array([[1,0],[S,1]])

A = (Ar.dot(Asc)).dot(Ash)

#A = numpy.array([[1,-1],[1,1]]) 

center = numpy.array([height/2,width/2])
b = center-A.dot(center)
b += numpy.array([0,0])

Ab = numpy.append(A,b.reshape(len(A),1),1)
print(Ab)
registImg = GrayImg.AFT(image,Ab)

plt.figure(figsize=(10,5))
plt.imshow(registImg , cmap='gray', vmin=0, vmax=255)
cv2.imwrite('Image\ReferenceImg.png',registImg)
```

# Vector Field

```python
nx = 10
ny = 15
hx = height//nx
hy = width//ny
px = []
py = []
tx = []
ty = []
vecField = 255*numpy.ones((height,width))
Ainv = numpy.linalg.inv(A)
for i in range(1,nx):
    for j in range(1,ny):
        px.append(i*hx)
        py.append(j*hy)
        #dstP = A.dot(numpy.array([i*hx,j*hy]))+b
        dstP = Ainv.dot(numpy.array([i*hx,j*hy])-b)
        tx.append(dstP[0])
        ty.append(dstP[1])
        
px = numpy.array(px)
py = numpy.array(py)
tx = numpy.array(tx)
ty = numpy.array(ty)

mt = ((tx-px)**2+(ty-py)**2)**(1/2)
maxt = numpy.max(mt)/numpy.min([hx,hy])

for k in range(0,(nx-1)*(ny-1)):
        cv2.arrowedLine(vecField,(py[k],px[k]),(floor(py[k]+0.5+(ty[k]-py[k])/maxt),floor(px[k]+0.5+(tx[k]-px[k])/maxt)),(0,0,0),1,8,0,0.2)
        
fig = plt.figure(figsize=(15,10))
fig.add_subplot(1,2,1)
plt.imshow(registImg , cmap='gray', vmin=0, vmax=255)
fig.add_subplot(1,2,2)
plt.imshow(vecField , cmap='gray', vmin=0, vmax=255)
```
